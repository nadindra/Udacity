### Opening and Reading into the data

First, import the necessary library that will used for this project

```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("titanic.csv")
```

check the dataset. We have 891 rows and 12 columns

```python
df
```

### Data Handling

There are missing value in column Ticket and Cabin. So we will drop the column because it canit add any value to the analysis

```python
df = df.drop(['Ticket','Cabin'], axis=1)
```

Drop the row that contains null

```python
df = df.dropna() 
```

The dataset is clean so we have 712 rows and 10 columns

```python
df
```

### Make a Graph

```python
# specifies the parameters of our graphs
fig = plt.figure(figsize=(10,5), dpi=1600) 
alpha=alpha_scatterplot = 0.2 
alpha_bar_chart = 0.55


ax1 = plt.subplot2grid((1,2),(0,0))          
df.Survived.value_counts().plot(kind='bar', alpha=alpha_bar_chart)
ax1.set_xlim(-1, 2)
plt.xlabel("Survival Rate")
plt.ylabel("Passenger")
plt.title("Distribution of Survival, (1 = Survived)")    
```

From the graph above, the Non-survival rate(0) was with a count of approximately 430. Where the Survived(1) was at approximately 300. It can be seen that the Non-survival was proportionately greater than survival rate

```python
plt.subplot2grid((1,2),(0,1))
plt.scatter(df.Survived, df.Age, alpha=alpha_scatterplot)
plt.ylabel("Age") 
plt.xlabel("Survival Rate")
plt.grid(b=True, which='major', axis='y')  
plt.title("Survival by Age,  (1 = Survived)")
```

From the scatter plot above, it can be seen that non-survival have mean age 40. Where as the survival have mean age 20. But the non-survival have maximum age in 70 where survival have 80. So it cannot be justified that the non-survival age have older group than survival. 

```python
ax3 = plt.subplot2grid((2,2),(0,2))
df.Pclass.value_counts().plot(kind="barh", alpha=alpha_bar_chart)
ax3.set_ylim(-1, len(df.Pclass.value_counts()))
plt.xlabel("Passenger")
plt.ylabel("Class")
plt.title("Class Distribution")
```

From the graph above, it can be seen that 3rd class had approximately 350 passengers. Where the 2nd class had approximately 170 passengers and 1st class had approximately 160 passengers

```python
plt.subplot2grid((2,3),(1,0), colspan=2)
df.Age[df.Pclass == 1].plot(kind='kde')    
df.Age[df.Pclass == 2].plot(kind='kde')
df.Age[df.Pclass == 3].plot(kind='kde')
plt.xlabel("Age")    
plt.title("Age Distribution within classes")
plt.legend(('1st class', '2nd class','3rd class'),loc='best') 
```

From the plot above, it can be seen that the age distribution of 3rd class had more young and child people than 2nd class. Where 3rd class had normal distribution with the maximum age is age of 30.

```python
ax5 = plt.subplot2grid((2,4),(1,1))
df.Embarked.value_counts().plot(kind='bar', alpha=alpha_bar_chart)
ax5.set_xlim(-1, len(df.Embarked.value_counts()))
plt.xlabel("Boarding Location")
plt.ylabel("Passenger")
plt.title("Passengers per boarding location")
```

From the graph above, it can be seen that passengers embark from S was with a count of approximately 550. Where C was with a count of approximately 130 and Q 50. It can be clearly justified that S is the most passengers embarked location.

### Exploratory Visualization

This section is to answer the questions above.
First, plot the bar graph with survival breakdown

```python
plt.figure(figsize=(6,4))
fig, ax = plt.subplots()
df.Survived.value_counts().plot(kind='barh', color="blue", alpha=.65)
ax.set_ylim(-1, len(df.Survived.value_counts())) 
plt.title("Survival Breakdown (1 = Survived, 0 = Died)")
```

Breakdown the graphy by gender

```python
fig = plt.figure(figsize=(18,6))


df_male = df.Survived[df.Sex == 'male'].value_counts().sort_index()
df_female = df.Survived[df.Sex == 'female'].value_counts().sort_index()

ax1 = fig.add_subplot(121)
df_male.plot(kind='barh',label='Male', alpha=0.55)
df_female.plot(kind='barh', color='#FA2379',label='Female', alpha=0.55)
plt.title("Who Survived? with respect to Gender, (raw value counts) "); plt.legend(loc='best')
ax1.set_ylim(-1, 2) 


ax2 = fig.add_subplot(122)
(df_male/float(df_male.sum())).plot(kind='barh',label='Male', alpha=0.55)  
(df_female/float(df_female.sum())).plot(kind='barh', color='#FA2379',label='Female', alpha=0.55)
plt.title("Who Survived proportionally? with respect to Gender"); plt.legend(loc='best')

ax2.set_ylim(-1, 2)
```

There is a clear visualization for men died and survived in value counts. And yes female had a great survival rate than men.

We go further by breakdown the graph with what class they were travelling in.

```python
fig = plt.figure(figsize=(18,4), dpi=1600)
alpha_level = 0.65


ax1=fig.add_subplot(141)
female_highclass = df.Survived[df.Sex == 'female'][df.Pclass != 3].value_counts()
female_highclass.plot(kind='bar', label='female, highclass', color='#FA2479', alpha=alpha_level)
ax1.set_xticklabels(["Survived", "Died"], rotation=0)
ax1.set_xlim(-1, len(female_highclass))
plt.title("Who Survived? with respect to Gender and Class"); plt.legend(loc='best')

ax2=fig.add_subplot(142, sharey=ax1)
female_lowclass = df.Survived[df.Sex == 'female'][df.Pclass == 3].value_counts()
female_lowclass.plot(kind='bar', label='female, low class', color='pink', alpha=alpha_level)
ax2.set_xticklabels(["Died","Survived"], rotation=0)
ax2.set_xlim(-1, len(female_lowclass))
plt.legend(loc='best')

ax3=fig.add_subplot(143, sharey=ax1)
male_lowclass = df.Survived[df.Sex == 'male'][df.Pclass == 3].value_counts()
male_lowclass.plot(kind='bar', label='male, low class',color='lightblue', alpha=alpha_level)
ax3.set_xticklabels(["Died","Survived"], rotation=0)
ax3.set_xlim(-1, len(male_lowclass))
plt.legend(loc='best')

ax4=fig.add_subplot(144, sharey=ax1)
male_highclass = df.Survived[df.Sex == 'male'][df.Pclass != 3].value_counts()
male_highclass.plot(kind='bar', label='male, highclass', alpha=alpha_level, color='steelblue')
ax4.set_xticklabels(["Died","Survived"], rotation=0)
ax4.set_xlim(-1, len(male_highclass))
plt.legend(loc='best')
```

So there are informarion who survived and died in the titanic tragedy.
